```python
# Reading dataframe
import pandas as pd

df = pd.read_csv('creditcard.csv')
df.head()
```

```python
# Unique Value counts
df.Class.value_counts()
```

```python
# Visualising The Target Variable
import seaborn as sns

sns.countplot(data = df, x = "Class")
```

```python
# Separating X and y 
# [Independent and Dependent variabele]
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
  
# split into 70:30 ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train.head()
```

```python
# Train  freq dist
sns.histplot(y_train )
```

```python
# test freq dist
sns.histplot(y_test)
```

# Scaling

**StandardScaler** scales the features to have a mean (average) of 0 and a standard deviation of 1. It transforms the data to follow a standard normal distribution.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
standardScalar_DF = pd.DataFrame(X_train_scaled, columns=df.columns[:-1])
standardScalar_DF.head()
```

**MinMaxScalar**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
MinMaxScalar_DF = pd.DataFrame(X_train_scaled, columns=df.columns[:-1])
MinMaxScalar_DF.head()
```
